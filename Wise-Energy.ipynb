```python
sel2_sql = '''select date(substr(epoch,1,10)) as ccyy_dt, substr(epoch,12,2),
    sum(irradiation_calc) 
    from api5_dv_vw api
    inner join 
    pyras pyr
    on api.fk_id_device = pyr.fk_id_device 
    where pyr.tilt != 0
    group by ccyy_dt, substr(epoch,12,2)
    having time(substr(epoch,12,8)) < '23:00:00'
    and time(substr(epoch,12,8)) > '03:00:00'
    '''

sel2_df= pd.read_sql(sel2_sql, db.engine)

sel2_df = sel2_df.groupby(by="ccyy_dt").agg(Irradiation=("sum(irradiation_calc)","sum")).reset_index(False)
sel2_df
```

```python
# THe pr is getting more than 100% and is missing for the last date. THere is a mismatch between the counnt of 
# records of the above 2 SQL queries

def find_pr(df):
    dc_capacity = 4.956
    return 100 * (df["Generation"] / (df["Irradiation"] * dc_capacity))

sel_all = pd.merge(sel1_df,sel2_df,on="ccyy_dt",how="inner")

pr_df = sel_all.apply(find_pr, axis=1) 
pr_df
```
